```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diwalidata = pd.read_csv(r"E:\DataSets\DataAnalysis\Diwali Sales Data.csv" , encoding='unicode_escape')
```

```python
diwalidata
```

```python
diwalidata = diwalidata.drop(['Status','unnamed1'], axis=1)
```

```python
diwalidata.shape
```

```python
diwalidata.head()
```

```python
diwalidata['Amount'].dtype
```

```python
pd.isnull(diwalidata).sum()
```

```python
diwalidata = diwalidata.dropna()
```

```python
pd.isnull(diwalidata).sum()
```

```python
gendervsamount = diwalidata.groupby(['Gender'])['Amount'].sum().sort_values()
```

```python
gendervsamount.plot.bar(color='Orange', title="AMOUNT SPEND BY GENDER")
```

### From the above visual we can see that most of the buyers are FEMALE. And purchasing power of females are greater than male

```python
diwalidata.head()
```

```python
age_groupvsamount = diwalidata.groupby(['Age Group'])['Amount'].sum().sort_values(ascending=False)
```

```python
age_groupvsamount
```

```python
age_groupvsamount.plot.bar(title="AMOUNT SPEND BY AGE GROUP")
```

### From the above graph we can tell that most amount spent by 26-35 years old people. 

```python
product_categoryvsorders = diwalidata.groupby(['Product_Category'])['Orders'].sum()
```

```python
product_categoryvsorders
```

```python
diwalidata['Product_ID'].count()
```

```python
product_categoryvsorders.plot.bar(figsize=(15,5), title="PRODUCT CATEGORY BY TOTAL ORDERS")
```

### Above Graph shows us Clothing & Apparel ordered most.

```python
productidvsorders = diwalidata.groupby(['Product_ID'])['Orders'].sum().nlargest(10).sort_values(ascending=False)
```

```python
productidvsorders.plot.bar(figsize=(15,5), title="TOP 10 PRODUCT SOLD")
```

### Above bar plot shows which 10 products ordered most.

```python
diwalidata.head()
```

```python
statevsorders = diwalidata.groupby(['State'])['Orders'].sum().sort_values(ascending=False)
```

```python
statevsorders
```

```python
statevsorders.plot.bar(figsize=(15,5),title='TOTAL ORDERS BY STATE')
```

### Above Graph shows that peoples of Uttar Pradesh ordered maximum number of product.
